## WHAT MAKES A FLIGHT MORE SATISFACTORY?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('airline_passenger_satisfaction.csv')
```

```python
df['Percentage of Delay'] = (df['Departure Delay']/df['Flight Distance'])*100
df.head()
```

```python
insatisfeitos = df[df['Satisfaction'] == "Neutral or Dissatisfied"]
```

```python
insatisfeitos.describe()
```

```python
df['Satisfaction'].unique()
```

```python
df['Class'].unique()
```

```python
df = pd.get_dummies(df, columns = ['Class'])
```

```python
df['Class_Business']
```

```python
dict_s = {'Satisfied': 0, 'Neutral or Dissatisfied': 1}
```

```python
df['reg_satisfaction'] = df['Satisfaction'].map(dict_s)
df.head()
```

```python
df.columns
```

```python
X = df[['Departure and Arrival Time Convenience', 'Ease of Online Booking',
       'Check-in Service', 'Online Boarding', 'Gate Location',
       'On-board Service', 'Seat Comfort', 'Leg Room Service', 'Cleanliness',
       'Food and Drink', 'In-flight Service', 'In-flight Wifi Service',
       'In-flight Entertainment', 'Baggage Handling', 'Class_Economy']]
y = df['reg_satisfaction']
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg = LogisticRegression()
```

```python
logreg.fit(X,y)
```

```python
logreg.predict(X)
```

```python
logreg.coef_
```

```python
from sklearn import tree
```

```python
clf = tree.DecisionTreeClassifier(max_depth=2)
```

```python
clf.fit(X,y)
```

```python
clf.predict(X)
```

```python
y_df = pd.DataFrame(y)
```

```python
y_df['predict'] = clf.predict(X)
```

```python
y_df.value_counts()
```

```python
y.head()
```

```python
plt.figure(figsize=(12,12))
tree.plot_tree(clf, feature_names = X.columns, fontsize = 12, filled = True);
```

```python
clf = tree.DecisionTreeClassifier(max_depth=3)
clf.fit(X,y)

plt.figure(figsize=(15,15))
tree.plot_tree(clf, feature_names = X.columns, fontsize = 12, filled = True);
```
